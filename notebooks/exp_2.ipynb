```python
import torch
from data import DensityMatrixDataset
from model import DensityNet
from metrics import quantum_fidelity, trace_distance

model = DensityNet()
model.load_state_dict(torch.load("outputs/model.pt"))
model.eval()

dataset = DensityMatrixDataset(50)

fidelities = []
trace_distances = []

with torch.no_grad():
    for x, true_L in dataset:
        pred_L = model(x.unsqueeze(0))

        L = pred_L.view(4, 4)
        rho_pred = L @ L.T
        rho_pred = rho_pred / torch.trace(rho_pred)

        true_L = true_L.view(4, 4)
        rho_true = true_L @ true_L.T
        rho_true = rho_true / torch.trace(rho_true)

        fidelities.append(quantum_fidelity(rho_true, rho_pred))
        trace_distances.append(trace_distance(rho_true, rho_pred))

print("Mean Fidelity:", sum(fidelities) / len(fidelities))
print("Mean Trace Distance:", sum(trace_distances) / len(trace_distances))
```
